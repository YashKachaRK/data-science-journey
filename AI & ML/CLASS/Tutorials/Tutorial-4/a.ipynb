```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
computer=pd.read_csv('data/computers.csv')
```

```python
computer.info()
```

```python
computer
```

```python
mean_value=computer['Minutes'].mean()
mean_value
```

```python
plt.scatter(computer['Units'],computer['Minutes'],color='blue')

plt.xlabel('Units')
plt.ylabel('Minutes')
```

```python
model0 = computer['Minutes'].mean()
```

```python
model1 = 10  + 12 * computer['Units']
```

```python
model2 = 6 + 18 * computer['Units'] 
```

```python
computer['Model_0']=model0
computer['Model_1']=model1
computer['Model_2']=model2
```

```python
computer
```

```python
fig, ax = plt.subplots()
ax.scatter(x='Units' , y ='Minutes', data = computer, label = 'actusl repair time')
ax.plot(computer['Units'] , computer ['Model_0'],color ='red' , label ='Model0')
ax.plot(computer['Units'] , computer ['Model_1'],color ='yellow' , label ='Model1')
ax.plot(computer['Units'] , computer ['Model_2'],color ='black' , label ='Model2')
```

```python
X = computer[['Units']]
y = computer['Minutes']
from sklearn.linear_model import LinearRegression
model3 = LinearRegression()
model3.fit(X,y)
print("Intercept:", model3.intercept_)
print("Coefficient:", model3.coef_)
```

```python
model3 = model3.intercept_ + model3.coef_[0] * computer['Units']
computer['Model_3']=model3
computer
```

```python
fig, ax = plt.subplots()
ax.scatter(x='Units' , y ='Minutes', data = computer, label = 'actusl repair time')
ax.plot(computer['Units'] , computer ['Model_0'],color ='red' , label ='Model0')
ax.plot(computer['Units'] , computer ['Model_1'],color ='yellow' , label ='Model1')
ax.plot(computer['Units'] , computer ['Model_2'],color ='black' , label ='Model2')
ax.plot(computer['Units'] , computer ['Model_3'],color ='green' , label ='Model3')
plt.xlabel('Units')
plt.ylabel('Minutes')
```
